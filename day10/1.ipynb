```python
class Student:
    def __init__(self, name, roll):
        self.name = name
        self.roll = roll
        self.marks = 0

    def add_marks(self, marks):
        self.marks = marks

    def display(self):
        print("----- Student Details -----")
        print("Name:", self.name)
        print("Roll No:", self.roll)
        print("Marks:", self.marks)

    def is_pass(self):
        if self.marks >= 40:
            print(self.name, "has PASSED ✅")
        else:
            print(self.name, "has FAILED ❌")


s1 = Student("Priyank", 101)
s2 = Student("Rahul", 102)

s1.add_marks(85)
s2.add_marks(35)


s1.display()
s1.is_pass()

print()

s2.display()
s2.is_pass()
```

```python
class Dog:
    def set_name(self, name):
        self.name = name

    def bark(self):
        print(f"{self.name} says: Woof Woof!")



dog1 = Dog()
dog1.set_name("Buddy")

dog2 = Dog()
dog2.set_name("Rocky")


dog1.bark()
dog2.bark()



```
