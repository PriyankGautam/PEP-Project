```python
class Teacher:
    def __init__(self, name, subject):
        self.name = name
        self.subject = subject

    def show_teacher(self):
        print("Teacher Name:", self.name)
        print("Subject:", self.subject)


class Student:
    def __init__(self, name, roll_no, bank_name, account_no):
        self.name = name              
        self.roll_no = roll_no        

        self.__bank_name = bank_name  
        self.__account_no = account_no  

    def show_student_name(self):
        print("Student Name:", self.name)

    def show_student_roll(self):
        print("Roll No:", self.roll_no)

    def show_bank_details(self):
        print("Bank Name:", self.__bank_name)
        print("Account No:", self.__account_no)



t1 = Teacher("Rajesh Sir", "Python")
t1.show_teacher()

print("---------------------")


s1 = Student("Priyank", 101, "SBI", "1234567890")

s1.show_student_name()
s1.show_student_roll()

print("---- Bank Details (Private) ----")
s1.show_bank_details()
```
